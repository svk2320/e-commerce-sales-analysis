```python
import sys
import os
sys.path.insert(0, os.path.abspath(".."))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.config import get_engine

engine = get_engine()
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
```

```python
query = """
    SELECT 
        DATE_TRUNC('month', o.order_purchase_timestamp) AS month,
        ROUND(SUM(op.payment_value)::numeric, 2)        AS revenue,
        COUNT(DISTINCT o.order_id)                       AS total_orders
    FROM orders_clean o
    JOIN order_payments op ON o.order_id = op.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY month
    ORDER BY month
"""

monthly = pd.read_sql(query, engine)
monthly["month"] = pd.to_datetime(monthly["month"])
monthly
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 10))

# Revenue trend
axes[0].plot(monthly["month"], monthly["revenue"], marker="o", color="steelblue", linewidth=2)
axes[0].set_title("Monthly Revenue Trend", fontsize=14)
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Revenue (BRL)")

# Orders trend
axes[1].plot(monthly["month"], monthly["total_orders"], marker="o", color="coral", linewidth=2)
axes[1].set_title("Monthly Orders Trend", fontsize=14)
axes[1].set_xlabel("Month")
axes[1].set_ylabel("Total Orders")

plt.tight_layout()
plt.show()
```

```python
query = """
    SELECT 
        TO_CHAR(o.order_purchase_timestamp, 'Day') AS day_of_week,
        EXTRACT(DOW FROM o.order_purchase_timestamp) AS day_num,
        COUNT(DISTINCT o.order_id)                   AS total_orders,
        ROUND(SUM(op.payment_value)::numeric, 2)     AS revenue
    FROM orders_clean o
    JOIN order_payments op ON o.order_id = op.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY day_of_week, day_num
    ORDER BY day_num
"""

daily = pd.read_sql(query, engine)
daily["day_of_week"] = daily["day_of_week"].str.strip()
daily
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.barplot(
    data=daily, x="day_of_week", y="total_orders",
    hue="day_of_week", palette="Blues_r", legend=False, ax=axes[0]
)
axes[0].set_title("Orders by Day of Week")
axes[0].set_xlabel("Day")
axes[0].set_ylabel("Total Orders")

sns.barplot(
    data=daily, x="day_of_week", y="revenue",
    hue="day_of_week", palette="Greens_r", legend=False, ax=axes[1]
)
axes[1].set_title("Revenue by Day of Week")
axes[1].set_xlabel("Day")
axes[1].set_ylabel("Revenue (BRL)")

plt.tight_layout()
plt.show()
```

```python
query = """
    SELECT 
        EXTRACT(MONTH FROM o.order_purchase_timestamp) AS month_num,
        TO_CHAR(o.order_purchase_timestamp, 'Month')   AS month_name,
        ROUND(SUM(op.payment_value)::numeric, 2)       AS revenue,
        COUNT(DISTINCT o.order_id)                     AS total_orders
    FROM orders_clean o
    JOIN order_payments op ON o.order_id = op.order_id
    WHERE o.order_status = 'delivered'
    GROUP BY month_num, month_name
    ORDER BY month_num
"""

seasonal = pd.read_sql(query, engine)
seasonal["month_name"] = seasonal["month_name"].str.strip()

plt.figure(figsize=(12, 5))
sns.barplot(
    data=seasonal, x="month_name", y="revenue",
    hue="month_name", palette="coolwarm", legend=False
)
plt.title("Revenue by Month (All Years Combined)", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Revenue (BRL)")
plt.tight_layout()
plt.show()
```
