# 03 — Revenue Analysis
**Phase 3 — Brazilian E-Commerce Analytics (Olist)**

Business questions answered:
- What is the total revenue, order count, and average order value?
- Which product categories generate the most revenue?
- Which payment types are most popular, and how do they differ in revenue?

```python
import sys
import os

# Add project root to path
sys.path.insert(0, os.path.abspath(".."))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.config import get_engine

engine = get_engine()

# Plot styling
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 5)
```

```python
query = """
    SELECT 
        ROUND(SUM(payment_value)::numeric, 2) AS total_revenue,
        COUNT(DISTINCT order_id)              AS total_orders,
        ROUND(AVG(payment_value)::numeric, 2) AS avg_order_value
    FROM order_payments
"""

summary = pd.read_sql(query, engine)
summary
```

```python
query = """
    SELECT 
        p.product_category_name_english AS category,
        ROUND(SUM(op.payment_value)::numeric, 2) AS revenue
    FROM order_payments op
    JOIN order_items oi ON op.order_id = oi.order_id
    JOIN products_clean p ON oi.product_id = p.product_id
    WHERE p.product_category_name_english != 'unknown'
    GROUP BY category
    ORDER BY revenue DESC
    LIMIT 10
"""

revenue_by_category = pd.read_sql(query, engine)
revenue_by_category
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=revenue_by_category,
    x="revenue",
    y="category",
    hue="category",
    palette="Blues_r",
    legend=False
)
plt.title("Top 10 Categories by Revenue", fontsize=14)
plt.xlabel("Revenue (BRL)")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
query = """
    SELECT 
        payment_type,
        COUNT(*)                                  AS total_orders,
        ROUND(SUM(payment_value)::numeric, 2)     AS total_revenue,
        ROUND(AVG(payment_value)::numeric, 2)     AS avg_value
    FROM order_payments
    GROUP BY payment_type
    ORDER BY total_revenue DESC
"""

payment_analysis = pd.read_sql(query, engine)
payment_analysis
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Orders by payment type
axes[0].pie(
    payment_analysis["total_orders"],
    labels=payment_analysis["payment_type"],
    autopct="%1.1f%%",
    startangle=90
)
axes[0].set_title("Orders by Payment Type")

# Revenue by payment type
axes[1].pie(
    payment_analysis["total_revenue"],
    labels=payment_analysis["payment_type"],
    autopct="%1.1f%%",
    startangle=90
)
axes[1].set_title("Revenue by Payment Type")

plt.tight_layout()
plt.show()
```
